```python
import json
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon
import contextily as ctx


# Read the data
#f = open('excel_instance.json')
f = open('small_instance.json')
instance_json = json.load(f)

clients_points = [Point(t[0], t[1]) for t in instance_json["instance"]["client_position"]]
depots_points = [Point(t[0], t[1]) for t in instance_json["instance"]["depot_position"]]

# Visualizing instance in a map
filename = 'data/perimetro/perimetro_wgs84.shp'
gdf = gpd.read_file(filename)
gs_clients = gpd.GeoSeries(clients_points)
gs_depots = gpd.GeoSeries(depots_points)


ax = gdf.plot(figsize=(10, 10), alpha=0.1, edgecolor="k")

gs_clients.plot(ax=ax, color='r', marker=".", markersize=10)
gs_depots.plot(ax=ax, color='b', marker=".", markersize=10)
ax.set_axis_off()
plt.show()

# Visualize instance information
demand_bins = list(range(5,21))
demands = instance_json["instance"]["demands"]
plt.hist(demands, bins=demand_bins,facecolor = '#2ab0ff', edgecolor='#169acf', linewidth=0.5)
plt.title('Distribucin de demandas')
plt.xlabel('Demanda') 
plt.ylabel('Cantidad') 
plt.show()

capacities = instance_json["instance"]["capacities"]
capacity_bins = list(range(min(capacities), max(capacities)+50, 25))
plt.hist(capacities, bins=capacity_bins,facecolor = '#2ab0ff', edgecolor='#169acf', linewidth=0.5)
plt.title('Distribucin de capacidades de depsitos')
plt.xlabel('Capacidad') 
plt.ylabel('Cantidad') 
plt.show()

print(max(capacities))
```

```python
from gap_instance import GAPInstance, GAPSolution, GAPChecker
from gap_ilp_solver import GAPIlpSolver

instance = GAPInstance()
solver = GAPIlpSolver()
checker = GAPChecker()

instance.from_json(instance_json)

# solve the model
solver.solve(instance)
solution = solver.get_solution()

# get the solution
ilp_result = solver.export_results()
ilp_result['instance_name'] = instance_json['name']

distances = checker.get_traveled_distances(instance, solution)
lfs = checker.get_load_factors(instance, solution)

#print(ilp_result)
```

```python
# Convertimos la instancia a formato del TP

output_file = open('real_instance','w')

output_file.write(str(instance.m) + " " + str(instance.n) + "\n")

for i in range(instance.m):
    line = ' '.join(map(str, instance.costs[i]))
    output_file.write(line + "\n")
    
for i in range(instance.m):
    line = ' '.join(map(str, instance.demands))
    output_file.write(line + "\n")

line = ' '.join(map(str, instance.capacities))
output_file.write(line + "\n")
output_file.close()
```
